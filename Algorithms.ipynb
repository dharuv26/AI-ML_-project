```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("vehiclesFinal.csv")
df=df.drop('id',axis=1)

df2=df.copy()
df.head()
```

```python
num_col=['year','odometer','long','lat']
cat_cols=['region','manufacturer','model','condition','cylinders','fuel','title_status','transmission','drive','size','type','paint_color']
```

```python
le=preprocessing.LabelEncoder()
df[cat_cols]=df[cat_cols].apply(le.fit_transform)

```

```python
df=df.sample(1000)
df.head(10)
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print(X.shape,y.shape)
```

```python

scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,confusion_matrix,r2_score,mean_squared_log_error

# def remove_neg(X_test,y_test,y_pred):
#     ind=[index for index in range(len(y_pred)) if(y_pred[index]>0)]
#     X_test=X_test[ind]
#     y_pred=y_pred[ind]
#     y_test=y_test[ind]
#     y_pred[y_pred<0]
#     return (X_test,y_test,y_pred)

def remove_neg(X_test, y_test, y_pred):
    mask = y_pred > 0  # Create a mask for positive values
    X_test = X_test[mask]
    y_pred = y_pred[mask]
    y_test = y_test[mask]
    return (X_test, y_test, y_pred)


def result(y_pred,y_test):
    # cm = confusion_matrix(y_test,y_pred)
    print("r2_score:", r2_score(y_pred,y_test))
    print("MSLE:", mean_squared_log_error(y_pred,y_test))
    print("MAE:" , mean_absolute_error(y_pred,y_test))
    print("MSE:" , mean_squared_error(y_pred,y_test))
    print("RMSE:" , root_mean_squared_error(y_pred,y_test))
    # sns.heatmap(cm,annot = True,xticklabels=data.target_names,yticklabels=data.target_names)
    # plt.show()
    return
```

```python
from sklearn.linear_model import LinearRegression

LR=LinearRegression()
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
X_test,y_test, y_pred=remove_neg(X_test,y_test,y_pred)

print("Coefficients: ",LR.coef_)
print()
print("Intercept: ",LR.intercept_)
print()
result(y_pred,y_test)
print()
print("Score: ", LR.score(X_test,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
result(y_pred, y_test)
print()
print("Score: ", logreg.score(X_test, y_test))
```
